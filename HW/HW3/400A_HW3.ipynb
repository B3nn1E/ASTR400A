```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

f = pd.read_csv(r'c:\Users\kietb\Downloads\Sun_Sim_Data\MESA-Web_Job_10022434412\profile8.data', delim_whitespace=True, header=4)
f
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(18, 5))  
ax1.plot(f['radius'], f['mass'])
ax1.set_xlabel('Radius ( R☉)')
ax1.set_ylabel('Mass (M☉))')
ax1.set_title('M(r) Graph')

ax2.plot(f['radius'], -f['pressure'], color = 'r')
ax2.set_xlabel('Radius ( R☉)')
ax2.set_ylabel('Negative Pressure (dyn/cm^2))')
ax2.set_title('- P(r) Graph')

ax3.plot(f['radius'], f['luminosity'], color = 'g')
ax3.set_xlabel('Radius ( R☉)')
ax3.set_ylabel('Luminosity(L⊙)')
ax3.set_title('L(r) Graph')


plt.show()
```

```python
fig, ax = plt.subplots() 
ax.plot(f['radius'], f['pressure'], label = 'P(r)')
ax.plot(f['mass'], f['pressure'],  label = 'P(m)')

ax.set_xlabel('Radius (R☉)/ Mass (M☉)')  
ax.set_ylabel('Pressure (dyn/cm^2)') 
ax.set_title('P(r) vs P(m)')  
ax.legend()  

plt.show() 
```
