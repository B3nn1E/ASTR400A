```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

star_norm = pd.read_csv(r'C:\Users\kietb\OneDrive\Desktop\Suffering\Undergrad\ASTR400A\Data\Sun_Sim_Data\MESA-Web_Job_10022434412\trimmed_history.data', delim_whitespace=True, header=4)
star_irr = pd.read_csv(r'C:\Users\kietb\OneDrive\Desktop\Suffering\Undergrad\ASTR400A\Data\Irregular_Sun_Sim_Data\MESA-Web_Job_10102434786\trimmed_history.data', delim_whitespace=True, header=4)
```

```python
# For normal star
norm_LogT = star_norm['log_center_T']
norm_LogL = star_norm['log_L']

# For irregular star
irr_LogT = star_irr['log_center_T']
irr_LogL = star_irr['log_L']

# Plot 
fig, ax = plt.subplots(figsize=(8, 6))
# For normal star
ax.plot (norm_LogT, norm_LogL, label = 'Star with Energy', color = 'red')

# For irregular star
ax.plot (irr_LogT, irr_LogL, label = 'Star without Energy', color = 'yellow')

# Invert the x-axis (HR diagrams go from high to low temp)
plt.gca().invert_xaxis()

# Labels and title
ax.set_xlabel('log(Effective Temperature) [K]')
ax.set_ylabel('log(Luminosity) [L⊙]')
ax.set_title('HR Diagram: 1M☉ Star Energy vs No Energy Evolutional Tracks')
ax.legend()

# Show the plot
plt.show()
```

```python

# Plot 
fig, ax = plt.subplots(figsize=(8, 6))
# For normal star
ax.plot (norm_LogT, norm_LogL, label = 'Star with Energy', color = 'orange')

# For irregular star
ax.plot (irr_LogT, irr_LogL, label = 'Star without Energy', color = 'green')

# Invert the x-axis (HR diagrams go from high to low temp)
plt.gca().invert_xaxis()

# Set up the constant radii
radii = [0.1, 0.25, 0.5, 0.75, 1]  # Radii in solar units

# Plot constant-radius lines using the relationship: L ∝ R^2 T^4
logT_values = np.linspace(3.5, 5, 100)  # Range of log(T) I see for these data
for R in radii:
    logL_values = 2 * np.log10(R) + 4 * logT_values  # Derive from L = 4pi*R^2*T^4*sigma
    ax.plot(logT_values, logL_values, '--', label=f'R = {R}R$_\odot$', color='gray')

# Labels and title
ax.set_xlabel('log(Effective Temperature) [K]')
ax.set_ylabel('Log(Luminosity) [L⊙]')
ax.set_title('HR Diagram: 1M☉ Star Energy vs No Energy Evolutional Tracks')
ax.legend()

# Show the plot
plt.show()
```

```python
# Calculate the average masss, radius, and density
average_mass = star_irr['mass'].mean()
average_radius = star_irr['radius'].mean()
average_density = average_mass / ((4/3)* np.pi * average_radius**3)

# Print the results
print(f"Average Mass: {average_mass}")
print(f"Average Radius: {average_radius}")
print(f"Average Density: {average_density}")
```
